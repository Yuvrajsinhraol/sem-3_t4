## super()

```python
class parent:
    def __init__(self):
        self.alter1=50
        self.alter2=60
class child(parent):
    def __init__(self):
        super().__init__()
        self.alter3=45
obj=child()
print(obj.alter3)
print(obj.alter1)
print(obj.alter2)
```

```python
class parent:
    def display(self ):
        print("hello parent")
class child(parent):
    def display(self):
        super().display()
        print("hello child")
obj=child()
obj.display()
```

* method resolution order(MRO)

- it demotes a way a programing language resolves a method or atribute .
- Mro defines the order in which the base classes are searched when executing method .
- First the method is searched within the class and then it follows the order we specified 
- while inheriteing.


```python
class A:
    def rk(self):
        print("prrint class a ")
class B(A):
    def rk(self):
        print("class B")
class C(B):
    def rk(self):
        print("class C")
class D(B,C):
    def rk(self):
        print("hello")
r=B()
r.rk()
    
```
