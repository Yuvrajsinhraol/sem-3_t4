## Bar Chart -bar()


```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array (['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y,color='darkblue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array (['A','B','C','D'])
y=np.array([3,8,1,10])
plt.barh(x,y,color='red')#barh gives horizontal
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array (['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y,color='red',width=0.4)#barh gives horizontal
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array (['A','B','C','D'])
y=np.array([3,8,1,10])
plt.barh(x,y,color='red',height=.4)#barh gives horizontal
plt.show()
```

## histogram  
-histogram is a graph showing frequency distribution.  
-it shows the no of observation within each given interval.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,1,2,3,3,3,3,4,4,5,7,8,8,9,10,10])
plt.hist(x,bins=5,orientation='horizontal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(1,100,size=(50))
print(x)
plt.hist(x,bins=[0,10,20,30,40,50,60,70,99])
plt.show()
```

## Area plot


```python
import matplotlib.pyplot as plt
import numpy as np
x=range(1,6)
y=[1,4,6,8,4]
plt.fill_between(x,y,color='orange',alpha=.5)
plt.plot(x,y,alpha=.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,4,4,5,6,4,3])
expense=np.array([6,1,6,8,3,10,9,6,4,5,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expense,color='red')
plt.fill_between(time,income,expense,where=(income>expense),color='green',alpha=.25,label='positive',interpolate=True)
plt.fill_between(time,income,expense,where=(income<expense),color='red',alpha=.25,label='negative',interpolate=True)
plt.legend()
```

```python
y=np.array([35,25,25,15])
lab=['Apple','Banana','Cherry','Dates']
plt.pie(y,labels=lab)
plt.show()
```

```python
y=np.array([35,25,25,15])
lab=['Apple','Banana','Cherry','Dates']
expl=[.3,0,1.0,.1]
plt.pie(y,labels=lab,explode=expl,colors=['red','yellow','maroon','sienna'],radius=2,autopct="%1.1f%%",shadow=True)
plt.legend(title="four fruits",loc=2)
plt.show()
```
